```python
import pandas as pd

# 1a. Calculate total units per segment
qty = (merged_data
       .groupby(['lifestage', 'premium_customer'])['prod_qty']
       .sum()
       .reset_index(name='TOTAL_QTY'))

# 1b. Count distinct customers per segment
cust = (merged_data
        .groupby(['lifestage', 'premium_customer'])['lylty_card_nbr']
        .nunique()
        .reset_index(name='NUM_CUSTOMERS'))

# 1c. Merge to compute average units per customer
avg_qty = pd.merge(qty, cust, on=['lifestage', 'premium_customer'])
avg_qty['AVG_UNITS_PER_CUST'] = avg_qty['TOTAL_QTY'] / avg_qty['NUM_CUSTOMERS']

print(avg_qty)
```

```python
import matplotlib.pyplot as plt

# Pivot for plotting
pivot_avg = avg_qty.pivot(index='lifestage',
                          columns='premium_customer',
                          values='AVG_UNITS_PER_CUST')

# Plot as bar chart
plt.figure(figsize=(10, 6))
pivot_avg.plot(kind='bar', stacked=False, colormap='tab20', ax=plt.gca())

plt.title('Average Chip Units Purchased per Customer by Segment')
plt.xlabel('Lifestage')
plt.ylabel('Avg Units per Customer')
plt.xticks(rotation=45, ha='right')
plt.legend(title='Premium Customer')
plt.tight_layout()
plt.show()
```

```python
# Aggregate sales and units by segment
sales = price.groupby(['lifestage','premium_customer']).agg(
    TOTAL_SALES=('TOTAL_SALES','sum'),
    TOTAL_UNITS=('prod_qty','sum')
).reset_index()

# Compute average price per unit
sales['AVG_PRICE_PER_UNIT'] = sales['TOTAL_SALES'] / sales['TOTAL_UNITS']
```

```python
import matplotlib.pyplot as plt

pivot_price = sales.pivot(index='lifestage',
                          columns='premium_customer',
                          values='AVG_PRICE_PER_UNIT')

plt.figure(figsize=(10,6))
pivot_price.plot(kind='bar', stacked=False, colormap='tab20', ax=plt.gca())
plt.title('Average Chip Price per Unit by Segment')
plt.xlabel('Lifestage')
plt.ylabel('Avg Price per Packet ($)')
plt.xticks(rotation=45, ha='right')
plt.legend(title='Premium Customer')
plt.tight_layout()
plt.show()
```

```python
import pandas as pd

price['AVG_PRICE_PER_UNIT'] = price['TOTAL_SALES'] / price['prod_qty']
midage_data = price[price['lifestage'] == 'Midage']
mainstream_midage = midage_data[midage_data['premium_customer'] == 'No']['AVG_PRICE_PER_UNIT']
premium_midage = midage_data[midage_data['premium_customer'] == 'Yes']['AVG_PRICE_PER_UNIT']
young_data = price[price['lifestage'] == 'Young Singles and Couples']
budget_young = young_data[young_data['premium_customer'] == 'No']['AVG_PRICE_PER_UNIT']
mainstream_young = young_data[young_data['premium_customer'] == 'Yes']['AVG_PRICE_PER_UNIT']
```

```python
from scipy import stats

t_stat_midage, p_value_midage = stats.ttest_ind(mainstream_midage, premium_midage, equal_var=True)
print(f"Mainstream Midage vs Premium Midage: t-statistic = {t_stat_midage}, p-value = {p_value_midage}")
t_stat_young, p_value_young = stats.ttest_ind(mainstream_young, budget_young, equal_var=True)
print(f"Young Singles and Couples vs Budget Young Singles and Couples: t-statistic = {t_stat_young}, p-value = {p_value_young}")
```
